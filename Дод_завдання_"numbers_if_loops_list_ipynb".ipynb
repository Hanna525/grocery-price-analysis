### **Додаткові задачі на роботу з числами та базові арифметичні операції**

1. Користувач вводить три числа. Обчисліть середнє арифметичне цих чисел.
2. Користувач вводить число. Перевірте, чи є воно парним або непарним.
3. Користувач вводить число і відсоток. Обчисліть відсоток від числа (наприклад, 15% від 200).
4. Користувач вводить число. Округліть його до найближчого цілого числа.
5. Користувач вводить два числа (початкова та кінцева ціна акції). Обчисліть відсоткову зміну.
6. Користувач вводить три числа (ціна покупки, ціна продажу, кількість акцій). Розрахуйте прибуток або збиток.
7. Користувач вводить три числа (початковий вклад, річна ставка у %, строк у роках). Розрахуйте суму на рахунку за формулою складних відсотків.
8. Користувач вводить два числа (загальна кількість товарів і кількість одиниць в одній упаковці). Визначте мінімальну кількість упаковок, необхідну для упаковки всіх товарів.
9. Користувач вводить чотири числа (дві точки на площині: x₁, y₁, x₂, y₂). Обчисліть відстань між цими точками за формулою евклідової відстані.
10. Користувач вводить суму доходу і суму витрат за місяць. Розрахуйте:

* коефіцієнт фінансової незалежності (дохід / витрати),
* відсоток доходу, що залишився після витрат,
* можливі заощадження за рік, якщо користувач зберігає незмінний рівень доходів і витрат.

### **Додаткові задачі на цикли та списки**

1. **Підрахунок загальної виручки.** У вас є список щоденної виручки магазину за тиждень (7 днів). Знайдіть загальну суму виручки.
   *Вхідні дані:* список із 7 чисел.
2. **Визначення кількості позитивних значень.** Дано список чисел, який містить прибутки та збитки компанії за кілька місяців. Порахуйте, скільки місяців компанія працювала з прибутком.
   *Вхідні дані:* список цілих чисел.
3. **Пошук максимального значення.** Дано список чисел, що містить температури за 10 днів. Знайдіть максимальну температуру.
   *Вхідні дані:* список із 10 чисел.
4. **Середнє значення.** Дано список чисел, що містить денну відвідуваність сайту за місяць (30 днів). Знайдіть середню відвідуваність.
   *Вхідні дані:* список із 30 цілих чисел.
5. **Підрахунок кількості парних чисел.** Дано список чисел, що містить ID клієнтів. Визначте, скільки ID є парними.
   *Вхідні дані:* список цілих чисел.
6. **Фільтрація даних.** Дано список чисел, що представляють ціни товарів в інтернет-магазині. Визначте, скільки товарів коштує дорожче 1000.
   *Вхідні дані:* список додатних чисел.
7. **Підрахунок загальної кількості товарів.** У магазині продаються товари кількох категорій. Дано список, де кожне число — кількість товарів у певній категорії. Знайдіть загальну кількість товарів.
   *Вхідні дані:* список цілих чисел.
8. **Пошук мінімального значення.** Дано список чисел, що містить затримки доставки замовлень (у днях) за місяць. Знайдіть мінімальну затримку.
   *Вхідні дані:* список із 30 чисел.
9. **Визначення відсотка бракованих товарів.** Завод випустив кілька партій товару, і серед них є браковані вироби (позначені як 0). Визначте відсоток браку.
   *Вхідні дані:* список із 1 та 0, де 1 — справний товар, 0 — брак.
10. **Порівняння двох магазинів.** Дано два списки, що містять щоденну виручку двох магазинів за тиждень. Визначте, скільки днів перший магазин заробив більше, ніж другий.
    *Вхідні дані:* два списки по 7 чисел.
11. **Середня зарплата.** Дано список зарплат працівника за 12 місяців. Визначте середню зарплату за рік і виведіть усі місяці, в яких зарплата була вищою за середню.
12. **Відсоткова зміна зарплати.** Дано список зарплат працівника за 12 місяців. Розрахуйте відсоткову зміну зарплати кожного місяця відносно попереднього.
13. **Премії працівників.** Дано список зарплат 5 працівників за рік (12 місяців для кожного). Знайдіть працівника, який отримав найбільшу премію (максимальна різниця між місяцями).
14. **Місяці з однаковою зарплатою.** Дано список зарплат працівника за 12 місяців. Знайдіть усі місяці, в яких зарплата була однаковою.
15. **Сума зарплат другого півріччя.** Дано список зарплат за рік. Розрахуйте загальну суму зарплат за друге півріччя (з липня по грудень).
16. **Нестабільна зарплата.** Дано список зарплат працівника за 12 місяців. Знайдіть три місяці поспіль, де зарплата змінювалась найбільше (сума змін між сусідніми місяцями максимальна).
17. **Порівняння зарплат двох працівників.** Дано два списки зарплат за 12 місяців для двох працівників. Виведіть список місяців, у яких зарплата першого працівника була вища.
18. **Накопичена зарплата.** Дано список зарплат за 12 місяців. Знайдіть місяць, у якому сумарна зарплата з початку року вперше перевищила 500 тисяч (або іншу суму).
19. **Коефіцієнт приросту зарплати.** Дано список зарплат працівника за 12 місяців. Знайдіть місяць із максимальним коефіцієнтом зростання зарплати порівняно з попереднім місяцем.
20. **Парні та непарні місяці.** Дано список зарплат за 12 місяців. Обчисліть загальну суму зарплат за парні та непарні місяці і порівняйте їх між собою.
